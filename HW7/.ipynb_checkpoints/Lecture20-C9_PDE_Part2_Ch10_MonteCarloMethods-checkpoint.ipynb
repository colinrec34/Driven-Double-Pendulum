## Chapter 9 - Partial differential equations (continued)
There is a little more to cover regarding *partial differential equations*. We covered *boundary value problems in the previous notebook. Now we return to the other type of differential equation problem, *initial value problems*. 

As a reminder of the difference, a boundary value problem specifies the values of the solution at a beginning point and ending point. We seek a solution that matches these. An example would be a projectile motion problem where we know the starting and ending points. We don't know the solution for position, or for the initial velocity. So part of the solution requires finding an initial velocity that provides a solution that matches the boundary values. 

Compare that to the same projectile motion problem, except as an initial value problem. In that case, we know the intitial position and initial velocity and must find a solution. The solution can tell us what the final position must be from the initial values.

As an example, let us consider the diffusion equation
$$
\frac{\partial \phi}{\partial t} = D\frac{\partial^{2} \phi}{\partial x^{2}}.
$$
D is the diffusion coefficient. It describes the motion of diffusing gases or liquids or the flow of heat. Therefore, it is sometimes called the *heat equation*.

The solution $\phi(x,t)$ is a function of both space and time. The form is similar to the Laplace equation which was solved in an earlier section of the textbook. But that was a boundary value problem. In this case, we do not know the value everywhere on the boundary. (Often we know the values on the boundaries of spatial dimensions but only the initial value for the time dimension. We don't know the values at the ending of our interval of time. Such problems are initial value problems.) Methods such as the relaxation method will not work. Instead we develop a *forward integration method*.

In one spatial dimension, x, a "grid" of points is really just a line of points a distance $a$ apart. We can use Equation 5.109 to represent the second derivative
$$
\frac{\partial^{2} \phi}{\partial x^{2}} = \frac{\phi(x+a,t) + \phi(x-a,t) - 2\phi(x,t)}{a^{2}}
$$
where we have just dropped the limit as $a \rightarrow 0$ to approximate the second derivative. Substituting into the diffuction equation we find
$$
\frac{\partial \phi}{\partial t} = \frac{D}{a^{2}} \left[ \frac{\phi(x+a,t) + \phi(x-a,t) - 2\phi(x,t)}{a^{2}} \right].
$$
In 1D the inital values 
A trick here is to now think of this from a different perspective. Instead of one solution at all gridpoints, consider this as separate ordinary differential equations, one at each grid point. These form a number of simultaneous ordinary differential equations of a type we solved in Chapter 8. The number depends on how many grid points we choose. A few hundred thousand can be solved on a computer. A few million becomes challenging. Beyond that requires a supercomputer.

The most common method to solve this is actually Euler's method. Even though it isn't very accurate (with errors of second order), the errors in the approximation of the second derivative, $\frac{\phi(x+a,t) + \phi(x-a,t) - 2\phi(x,t)}{a^{2}}/a^{2}$ also has errors of second order. Why waste time using a more accurate Runge-Kutta method if you will still have second-order errors from the numerical approximation of the derivative? There's no point.

To be clear, we know of a system of many, simultaneous, ordinary diffrential equations in *time*. One at each spatial grid point. Let us review Euler's method. The right side of our equations, we call $f(x,t)$ so our differential equations are 
$$
\frac{d \phi}{dt} = f(x,t)
$$
We Taylor expand the solution, $\phi(x,t)$, about t and obtain
$$
\phi(t+h) \approx \phi(t) + h\frac{d \phi}{dt} = \phi(t) + h f(\phi,t).
$$
For our diffusion equation we write an equivalent equation at each grid point, x.
$$
\phi(x,t+h) = \phi(x,t) + h\frac{D}{a^{2}} \left[ \frac{\phi(x+a,t) + \phi(x-a,t) - 2\phi(x,t)}{a^{2}} \right].
$$
If we know the value at every grid point (x, x+a, x-a, etc.) at t=0 (or any single value of t, really). We can use this equation to compute the value at each grid point at time step, t+h, later. Just repeat the process to compute the solution at further and further (n) time steps $(t + n \ h)$ into the future. This is the *forward-time centered-space* method, or FTCS method, for solving partial differential equations.

### Example 9.3 Heat Equation
Let's implement this for the following problem. A steel container, with 1 cm thick walls, has a uniform temperature of 20 degrees Celcius. It is placed in a bath of cold water at 0 degrees Celcius and filled with hot water at 50 degrees Celcius. (There is a figure an illustration on page 420 of the textbook.)

We wish to compute the temperature profile of the steel as a function of distance x from the hot side (x=0) to the cold side (x=1 cm), and also as a function of time. If we consider the base of the container as being arbitrarily wide, we may assume that the temperature profile is the same at each point. (It only changes in the direction from cold to hot, not parallel to the surface.) We also assume the cold and hot water do not change temperature significantly.

Thermal conduction is described by the diffusion equation
$$
\frac{\partial T}{\partial t} = D \frac{\partial^{2} T}{\partial x^{2}}
$$
where x is the position in the direction from cold to hot. We divide the thickness into 101 grid points across the 1 cm thickness of the container wall (the grid spacing is 0.1 mm). This will give one grid point at each of the surfaces. The initial temperatures at the first and last grid points are T=50 and T=0 degrees Celcius, respectively. All intermediate points are T=20 degrees Celcius. The temperature will gradually diffuse from hot to cold over time until a smooth profile is obtain. Finally, for steel, the *diffusion coefficient* is $D=4.25$ x $10^{-6}$ $m^{2}s^{-1}$.

In the following cell, code implements the FTCS method and plots the Temperature profile at t = 0.01 s, 0.1 s, 0.4 s, 1 s, and 10 s, on the same graph.

```python
from numpy import empty
from pylab import plot,xlabel,ylabel,show

# Constants
L = 0.01      # Thickness of steel in meters
D = 4.25e-6   # Thermal diffusivity
N = 100       # Numer of divisions in grid
a = L/N       # Grid spacing
h = 1e-4      # Time-step
epsilon = h/1000

Tlo = 0.0     # Low temperature in Celcius
Tmid = 20.0   # Intermediate temperature in Celcius
Thi = 50.0    # Hi temperature in Celcius

# These are the points in time we will plot our solutions.
t1 = 0.01
t2 = 0.1
t3 = 0.4
t4 = 1.0
t5 = 10.0
tend = t5+epsilon

# Create arrays
T = empty(N+1,float)
T[0] = Thi         # Set initial hi temperature edge
T[N] = Tlo         # Set initial low temperature edge
T[1:N] = Tmid      # Set all other points to intermediate temperature initially
Tp = empty(N+1,float)
Tp[0] = Thi        # Set initial values of our temperature profile, Tp
Tp[N] = Tlo

# Main loop
t = 0.0
c = h*D/(a*a)
while t<tend:
    
    # Calculate the new values of T
    Tp[1:N] = T[1:N] + c*(T[2:N+1] + T[0:N-1] - 2*T[1:N])  # This comes from the numerical second derivative Eq. 5.109
    T,Tp = Tp,T                                      # Switch values, new solutions fed back on the right side next time
    t += h
    
    # Make plots at the given time steps
    if abs(t-t1)<epsilon:
        plot(T)
    if abs(t-t2)<epsilon:
        plot(T)
    if abs(t-t3)<epsilon:
        plot(T)
    if abs(t-t4)<epsilon:
        plot(T)
    if abs(t-t5)<epsilon:
        plot(T)

xlabel("x (from hot to cold)")
ylabel("T (Celcius)")
show()
```

After setting up the solution, notice that we are computing a solution, feeding it back in on the right side the next time when we also increasing the time-step by h. Each time we are getting the entire profile, represented at each grid point in x, but always moving forward in time, t. It is an initial value problem because we are computing our value forward in time with no end point (boundary) specified.

Notice this program compares "abs(t-t1)<epsilon" which is because the values of t are floating-point numbers and they are subject to numerical error. We generate a plot when the time, t, is within a small range we can adjust as needed.

A second trick (in Python), that we could implement, would remove the "for" loop. Since Python can do arithmetic on entire arrays, we could replace

for i in range(1,N):
    Tp[i] = T[i] + c*(T[i+1] + T[i-1] - 2*T[i])

with

Tp[1:N] = T[1:N] + c*(T[2:N+1] + T[0:N-1] - 2*T[1:N])

This will be a huge increase in speed because Python can do this array calculation almost as fast as a calculation with a single variable!

# Chapter 10 - Random processes and Monte Carlo methods

Many processes we observe are random in nature. Radioactive decay is just one example. On a computer, methods have been developed to study such processes. True *random numbers* can be approximated by *pseudorandom* numbers. As we will see, these appear random, though they come from deterministic formula. We can reproduce them as long as we start from the same initial value (seed value), as we shall see.

### Random number generators

Random number generators have been developed and they have different quality, complexity and speed. Which one to use in a particular case is a balance of these considerations. To understand these generators, we begin by looking at a simple method of generating random numbers: the *Linear Congruential Random Number Generator*. It is not of high quality, but it is simple and will demonstrate the main points. It is fine for quick, simple calculations but not suitable for professional physics analysis.

This method uses the equation
$$
x^{'}=(ax+c)\text{mod}(m)
$$
where $a=1 \ 664 \ 525$ $c = 1 \ 013 \ 904 \ 223$, and $m = 4 \ 294 \ 967 \ 296$. Random numbers are generated by selecting an initial value of x (*seed* value), computing a value of $x^{'}$ which is the next random number, and feeding this value of $x^{'}$ back in for $x$ on the right side. Repeat N times to generate N random numbers (or rather *pseudorandom* numbers). Here is code to do this (From the textbook, p. 445). It generates a plot with the first 100 random numbers.

```python
from pylab import plot,show

N = 100
a = 1664525
c = 10-13904223
m = 4294967296
x = 1
results = []

for i in range(N):
    x = (a*x+c)%m
    results.append(x)

plot(results,"o")
show()
```

There are four things to observe (that are common to random number generators, in general).
1. It is not really random. Run it several times and notice the points on the graph don't change at all. No change the starting value of x to 2. Does it change? We can reproduce the same set of "random" numbers by using the same *seed* value to start. In physics we need values that are "random enough" and much work has been put into developing good random number generators.
2. The numbers in this case are positive or zero and always less than $m$,  that is $(0 \leq x < m-1)$. Often we want values in the range $0 \leq r < 1$. In this example, we could just divide all the numbers generated by m. 
3. The values of a, c and m were chosen for a specific purpose. Some choices do not give random numbers. (If c and m are both even only even numbers are generated which is not random.) This algorithm has been studied and the particular values used are known to give fairly random numbers.
4. While a, c and m are chosen, different sequeces of random numbers can be obtained by setting a different starting value of x (a different *seed* value). As you saw in 1.

The reason this *linear congruent random number generator* is a bad random number generator is that there are correlations between successive numbers. Such correlations can affect results of your computational studies but be very difficult to detect.

Many random number generators exist. In physics, one of the most commonly used generators is called the *Mersenne twister*, a "generalized feedback shift-register generator." We will not go into what that means. It is rather complicated to code and understand, but we can just import it from the **random** package in **numpy**. It can be accessed with four functions:

random() - Gives random **floating-point** numbers in range $0 \leq r < 1$.  
random(n) - Gives random **integers**, $0 \leq r < n-1$.  
random(m,n) - Gives random integers, $m \leq r < n-1$.  
randrange(m,n,k) - Gives random integers in range $m \leq r < n-1$ in steps of k.  

These all return random values using the **Mersenne twister** and are adequate for serious physics calculations. They are not the best random number generators but they are a compromise between quality and speed. They are adequate for most physics studies (maybe not for cryptography, however).

Notice with these functions return a different value each time...

## Section 10.1.4 - Probabilities and biased coins
So far all the random number functions have generated random numbers that have a uniform distribution. The histogram created in the next cell shows the distribution of one of the built in functions. The distribution is flat. Any number in the range is equally likely. This is to be compared some other distribution



```python
#  Uniform distribution of standard random number functions
from random import randrange
from pylab import plot,show
from numpy import array,zeros,empty,mean

rand_dist = zeros(100,float)    # A histogram to count random numbers in each bin
rand_axis = zeros(100,float)    # An axis with one entry for each bin in our histogram
rand_min = 0                    # Lower limit of our random number range
rand_max = 10                   # Upper limit of our random number range
Nbins = 100                     # Number of bins in our random number distribution
bin_width = (rand_max-rand_min)/(Nbins) # Compute the bin width

# create an array, one element for each bin, containing the value of the lower edge of the bin
for i in range(len(rand_axis)):    
    rand_axis[i] = rand_min + (i)*bin_width    
print("bin width = ",bin_width)  # verify the bin width

N=100000                       # Number of random numbers to generate
for i in range(N):
    z = random()*10            # random floating-points numbers in range z=0 to z=10
    for j in range(len(rand_axis)+1):
        
        # add up the number of random numbers in each bin (should be uniform)
        if (z >= (rand_axis[j]) and z < (rand_axis[j]+bin_width)):
            rand_dist[j] += 1
            break
        
for i in range(len(rand_axis)):    
    rand_axis[i] += (0.5)*bin_width    # shift to bin centers for plotting
rand_dist /= N                         # Normalize so the area under the distribution is 1.

print(mean(rand_dist), sum(rand_dist)) # Compute mean and sum, should be 0.1 per bin, and normalized to 1

plot(rand_axis,rand_dist)
show()
```

Often we need to investigate events that occur with unequal probabilities. For instance what if a process only happens $1/5$ of the time. We can generate random numbers but control what happens using if statements...

```python
from random import random
if random()<0.2:
    print("heads")
else:
    print("tails")
```

Run this a few times and see that you get tails most (~80%) of the time. 

This is like a biased coin. It could be some other process. If we want the event to vary on a case-by-case basis, but with uneven probability, we can use a trick like this. We are generating random numbers to allow the case to vary, but imposing the unequal probability of the events. Usually, we can simply generate random numbers in the range 0 to 1 and split the interval as we want: $r<0.2$ or $r \geq 0.2$ in this case.

In more complicated situations, we can impose a know (or expected) probability distribution. Radioactive decay is a perfect example of a random process that follows a non-uniform distribution. In fact, it follows an exponential probability distribution. We can still use random numbers from a uniformly distribution but compare them to the probability of decay in each "time" step of our simulation. The probability is between 0 and 1.

For decay of $^{208}Tl$ (thallium 208) to $^{208}Pb$ (lead 208) with a half-life of 3.053 minutes, we will consider $N(0)=1000$ atomes initially. The number of Thallium atoms decreases exponentially according to
$$
N(t) = N(0)2^{-t/\tau}.
$$
At any time, t, the *fraction* of thallium atoms that remain is $N(t)/N(0)=2^{-t/\tau}$ so the probability (fraction) that have decayed is
$$
p(t) = 1-2^{-t/\tau}
$$
and the number of lead is found by multiplying $N(0)p(t)$.

The following code implements Example 10.1 (or file decay.py from the textbook website).

```python
from random import random
from numpy import arange
from pylab import plot,xlabel,ylabel,show

# Constants
NTl = 1000             # Number of thallium atoms
NPb = 0                # Number of lead atoms
tau = 3.053*60         # Half life of thallium in seconds
h = 1.0                # Size of time-step in seconds
p = 1 - 2**(-h/tau)    # Probability of decay in one step
tmax = 1000            # Total time

# Lists of plot points
tpoints = arange(0.0,tmax,h)
Tlpoints = []
Pbpoints = []

# Main loop
for t in tpoints:
    Tlpoints.append(NTl)
    Pbpoints.append(NPb)
    
    # Calculate the number of atoms that decay
    decay = 0
    for i in range(NTl):
        if random()<p:   # random numbers are uniformly distributed, but decay probability during step selects what happens
            decay += 1
    NTl -= decay
    NPb += decay

# Make the graph
plot(tpoints,Tlpoints)
plot(tpoints,Pbpoints)
xlabel("Time")
ylabel("Number of atoms")
show()
```

If we were just plotting the distributions $N(t)$ for thallium and $N_{Pb}(t) = N(0)p(t)$ for lead these would be perfectly smooth functions. However, we are generating these values randomly drawn from exponential distributions. So we don't get smooth curves, we get wiggles, similar to if we were to measure values in an experiment.

## Non-uniform probability distributions
Let's reconsider the radioactive decay problem from before. The decay follows a non-uniform (exponential) distribution. It would be easier if we could just generate random numbers that followed this distribution rather than a uniform one. The most common method to do this is the *transformation method*.

First, let's consider a probability density, $p(x)$ which is the probability of something happening in the interval $x$ to $x+dx$. (It's a probability *density* because the probability this function multiplied by the interval dx. In other words $p(x)dx$ is the probability of the event in the interval dx. This is the same probability of generating a value z in an interval dz. By definition
$$ 
p(x)dx = q(z)dz
$$
where the value of $x$ is related to the value of $z$ by x=x(z). Consider integrating these...
$$
\int_{-\infty}^{x(z)} p(x^{'} dx^{'} = \int_{0}^{z} dz^{'} = z.
$$
*If* we can do the interval on the left it will give us the transformation, x(z). We could convert uniform random numbers z to random numbers distributed according to the distributions $p(x)$, simply using the function x(z). Let's look at an example where
$$
p(x) = \mu e^{-\mu x}.
$$
where $\mu$ in front is a normalization constant. (This is the exponential distribution for radioactive decay with $\mu=ln \ 2/\tau$. Doing the integral
$$
\mu \int_{0}^{x(z)} e^{-\mu x^{'}} dx^{'} = 1 - e^{-\mu x} = z.
$$
which gives 
$$
x = -\frac{1}{\mu}ln(1-z)
$$
Generate random numbers, z, that are uniformly distributed from 0 to 1, using the standard, built-in random functions. If you compute values of x using this equation and you will have exponentially distributed random numbers x.

To summarize the *transformation method*, if we know the probability density, $p(x)$, we can compute a formula that converts uniform random numbers to random numbers that follow the probability density $p(x)$. 


## Section 10.1.6 - Gaussian random numbers
Because Gaussian distributions are encountered so frequently, let us look a bit at this case briefly. The calculation is outlined in Section 10.1.6 in detail but we will look at the main points here. Just as in the previous section for exponential distribution we wish to find a formula $x(z)$ that will convert uniform random numbers z to Gaussian random numbers x. We wish to generate random numbers drawn from a Gaussian (or normal) probability density distribution:
$$
p(x) = \frac{1}{\sqrt{2 \pi \sigma^{2}}} e^{-\frac{x}{2 \sigma^{2}}}
$$
where $\sigma$ is the standard deviation or width of the distribution and the factor in front is a normalization constant. Then
$$
p(x) = \frac{1}{\sqrt{2 \pi \sigma^{2}}} \int_{-\infty}^{x} e^{-\frac{x}{2 \sigma^{2}}} dx
$$
*would* give us the distribution $x(z)$ we want. The problem is we can not solve this integral. **This is a common problem in physics and math and there is a standard trick to get around this.** The trick is do the problem in two dimensions x and y. Convert to polar coordinates r and $\theta$. The book does this, but the key is the limits of integration for r are 0 to r, whereas the integral over x was $-\infty$ to x. The integral over $\theta$ is just $2\pi$. 

This process results in the equation
$$
r = \sqrt{-2*\sigma^{2} ln(1-z)}
$$
If you generate uniform random numbers z, you can compute Gaussian random numbers r. Combine these with uniform random numbers $\theta$ from 0 to $2\pi$ and you get Gaussian random numbers
$$
x = r \cos{\theta} \qquad \qquad y = r \sin{\theta}
$$
You get two Gaussian random numbers in Cartesion coordinates.

Textbook Example 10.2, illustrates an application where Gaussian distributed random numbers are useful. Rutherford scattering is a typical problem in any graduate classical mechanics course. You find that the scattering angle for helium nuclei off of an atom (gold atom in this case) is given by
$$
\tan{\frac{1}{2}\theta} = \frac{Ze^{2}}{2 \pi \epsilon_{0} E b}
$$
where Z is atomic number of nucleus, e is electron charge, $\epsilon_{0}$ is the permittivity of free space, E is kinetic energy of the incident particle and b is the impact parameter. (You may refer to figure 10.3 for an illustration of the problem.)

What Rutherford discovered was some of the helium nuclei bounced back! This was surprising because he did not yet know there was a thing called a nucleus with a huge, positive charge that could repel even these fast moving helium nuclei if it was nearly a head on collision. The particles that "bounce back" we will define as having scattering angles with $\theta=90$ degrees or more. Substituting $\theta = 90$ degrees, we find this happens when the impact parameter is smaller than
$$
b = \frac{Ze^{2}}{2 \pi \epsilon_{0} E}
$$.

In the cell below, enter the code from Example 10.2 for Rutherford scattering. This is a common problem assigned in graduate classical mechanics (at least solving the formula involved). Here we use the formula to learn some more about scattering.

```python
# Rutherford scattering, Example 10.2
from math import sqrt, log, cos, sin, pi
from random import random

#Constants
Z = 79
e = 1.602e-19
E = 7.7e6*e
epsilon0 = 8.854e-12
a0 = 5.292e-11
sigma = a0/100
N = 1000000

#Function to generate two Gaussian random numbers
def gaussian():
    r = sqrt(-2*sigma*sigma*log(1-random()))
    theta = 2*pi*random()
    x = r*cos(theta)
    y = r*sin(theta)
    return x, y

#Main program
count = 0
for i in range(N):
    x, y = gaussian()
    b = sqrt(x*x+y*y)
    if b<Z*e*e/(2*pi*epsilon0*E):
        count += 1
        
print(count, "particles were reflected out of",N)
```

## 10.2 - Monte Carlo integration
This is a neat trick to do integrals. For many integrals it is terrible and any appropriate method (trapezoidal, Simpson's, etc) from before will work better. However, if the function has pathologies, such as rapid fluctuations or divergences (infinities), or if the integral is over some high number of dimensions, it may be the only realistic method.

First, let's do a very common exercise. We will compute $\pi$ using random numbers. The area of a square of side $a=2$ is $A=a^{2}=4$. The area of a circle of radius $r=1$ is $A=\pi(1)^{2} = \pi$. The ratio  $A_{square}/A_{circle}=\pi/4$. Imagine the circle of radius $r=1$ inscribed in the square of side $a=2$. If we generate random points (x,y) with x and y each in the range 0 to 1? All are in the square, but only some fraction are in the area of the circle. The number in the circle have $r=\sqrt{x^{2}+y^{2}} < 1$. 

The reason this works, is that the probability a random number will be in the circle is proportional to the area of the circle. One hundred percent are in the square, because we choose all of the random points to be in the square.

```python
# Calculating pi
import math
N = 10**6
Ncircle = 0
for i in range(N):
    x = 2.0*random()-1.0
    y = 2.0*random()-1.0
    r = math.sqrt(x**2 + y**2)
    if r<1:
        Ncircle += 1

print("The ratio of points in the circle to points in the square is",4.0*Ncircle/N)
```

You have just computed the area of a circle of radius 1 simply by generating random numbers and counting the fraction in the region of interest versus the total number. But finding the area is exactly what we do when we integrate a function. We find the area between the curve and an axis.

Consider a similar problem using the textbook example performing the integral
$$
I = \int_{0}^{2} \sin^{2}\left[ \frac{1}{x(2-x)} \right] dx.
$$
This oscillates *infinitely* fast near x=0 and x=2. The entire curve fits inside the rectangle with $0<y<1$ and $0<x<2$ which has area 2x1=2. The area under the curve is **must** be less than two! None of the earlier methods (trapezoidal, Simpson's, Guassian quadrature) will work due to the infinitely fast oscillations.

What if we do the same thing as with the circle? Generate random points (x,y) inside this rectangle, count how many are "below" the curve and divide by the total number of points. The ratio approximates the value of the integral, I, divided by the area of the rectangle, A (=2x1 in this case). Therefore,
$$
\frac{k}{N} = \frac{I}{A} \qquad \text{or} \qquad I = \frac{kA}{N}.
$$

```python
# MC integration

from math import sin
from random import random

def f(x):
    return (sin(1/(x* (2-x))))**2

N = 10000 
count = 0
for i in range (N): 
    x = 2*random()
    y = random()
    if y<f(x):
        count += 1
I= 2*count/N 
print (I)
```

The problem with Monte Carlo integration is it does not give very accurate answers for the computational effort. But if it works when other methods do not it is better than nothing. And there are several reasons it may be the best method (speed, pathologies of the function that other methods cannot handle). 

First, let us compare the accuracy of the methods for the computational effort (number of steps basically). The textbook discusses in a bit more detail the different methods, but here we will just summarize.

|    Method    | Error on integral proportional to |
| --- | --- |
| Monte Carlo  | 1/$\sqrt{N}$                      |
| Trapezoidal  | 1/$N^{2}$                         |
| Simpson's    | 1/$N^{4}$                         |

If we increase the number of steps or random points by a factor of 100, the Monte Carlo integral improves by a factor of 10, the integral using the Trapezoidal method improves by a factor of 10 thousand, and the integral using Simpson's method improves by 10 million times. Clearly, if these other methods will work, they are preferable.

### 10.2.1 The mean value method
The integral of a function on an integral a to b is equal to the integral of its average value in that integral. But that is just the area of a rectangle, $(b-a)<f(x)>$, the average value $<f(x)>$ times the interval (b-a). The fundamental formulafor this *mean value method* is given by
$$
I \approx \frac{(b-a)}{N} \sum_{i=1}^{N} f(x_{i})
$$
This introduces a slight improvement (though not much) over the previous Monte Carlo implementation.

### 10.2.2 Integrals in many dimensions
Here is an example where Monte Carlo integration becomes necessary. Recall that integrating in two dimensions requires computing values on a two dimensional grid of points. This requires two nested *for* loops. In three dimensions, we need a 3D grid and three nested *for* loops. In four dimensions, if we have only 100 points for each dimension of the grid this is $100^{4}$ = 100 million integration points. Much beyond that and our calculation will become quite slow. And this is for only 100 points per side! The integral gets more accurate with smaller grid spacing.

This is a place were Monte Carlo integration becomes necessary. It finds applications in financial mathematics used to predict values of portfolios of stocks and bonds. Wall street firms hire a lot of physicists and mathemeticians to do financial analysis.

### 10.2.3 Importance sampling (pathological functions and infinite domains)
Some functions have divergences that may cause other integrations to fail. We can use a Monte Carlo approach called importance sampling to integrate such functions numerically. The textbook in Section 10.2.3 goes through the derivation in some detail. The formula they arrive at is a generalization of the *mean value method* above. 

Consider the integral
$$
I = \int_{0}^{1} \frac{x^{-1/2}}{e^{x}+1} dx
$$

This integral is finite but has diverges at x=0. If we try the mean value method above, the value of $f(x_{i})$ in the sum becomes very, very large as $x_{i} \rightarrow 0$. This dominates the sum. Depending on the values of $x_{i}$ we use, there can be large variations in the integral we compute. The estimate is not reliable.

The way around this is a method called *importance sampling*. Instead of using uniformly distribute values of $x_{i}$ we us values from a non-uniform distribution. In the integral above, we can choose the weight function 
$$
w(x) = x^{-1/2}
$$
and draw points $x_{i}$ according to a probability density function
$$
p(x) = \frac{w(x)}{\int_{a}^{b} w(x) dx}
$$
This is the same process we used to draw points from an exponential function earlier.

The textbook arrives at the **fundamental formula for importance sampling**
$$
I = \frac{1}{N} \sum_{i=1}^{n} \frac{f(x_{i})}{w(x_{i})} \int_{a}^{b} w(x) dx
$$
Notice what happens if $w(x)=1$ (and so $w(x_{i})=1$ also). The integral evaluates to $(b-a)$. The integral reduces to the *mean value method* formula. That method is a special case of importance sampling with equal weights.

To summarize the benefits of this approach, we can choose a weight function that removes the pathologies as we did in this example. The advantage is we can remove those pathologies and perform the integral. The disadvantage is that we must draw sample points from a non-uniform distribution which is a more complicated to figure out and implement. 

One more application of importance sampling is integrating over an infinite range, $x=0$ to $\infty$. The mean value method will not work because we would need to choose an infinite number of sample points, $x_{i}$, to sample an infinite range. Instead we can choose non-uniformly distribute points to weight our function using a distribution with infinite range. We already did this in the exponential distribution for radioactive decay. With our weight function $w(x)=e^{-x}$ our integral would be
$$
I = \frac{1}{N} \sum_{i=1}^{N} e^{x_{i}} f(x_{i}) \int_{0}^{\infty} f(x) dx = \frac{1}{N} \sum_{i=1}^{N} e^{x_{i}}f(x)
$$
We are essentially weighting the points more heavily when they are in a region with fewer points.

Other methods may still work over infinite ranges. However, if they have pathologies, those methods may again fail and this approach provides another way to approach the integral.
