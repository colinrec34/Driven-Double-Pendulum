````python
with open("sunspots.txt") as file:
    data = [float(i) for i in file.read().split()]
#print(data)
col1, col2 = data[::2], data[1::2]
array = np.array([col1, col2], dtype='int')
print(array)
````

**b)**

````python
array = array[:, :1000]
print(array)
````

**c)**

````python
X = array[0]
Y = array[1]
def runningavg(array, k, r):
    sum = 0
    for m in range(k-r, k+r+1):
        sum += array[m+r]
    return sum/(2*r+1)
running_avg = [runningavg(Y, i, 5) for i in Y]
print(running_avg)
````

# HR diagram example of scatterplot()

**a)** Copy code from hrdiagram.py and download the stars.txt file

**b)**: Reproduce Figure 3.4 in the text.

````python
from pylab import scatter,xlabel,ylabel,xlim,ylim,show
from numpy import loadtxt
data = loadtxt("stars.txt",float)
x = data[:,0]
y = data[:,1]
scatter(x,y)
xlabel("Temperature")
ylabel("Magnitude")
xlim(0, 13000)
ylim(-5,20)
show()
````
